```python
#install and then import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(1,201,2)
y = np.random.randint(1,10,100) #random.randiant use for take random number
```

```python
y
```

<h2 style="border:2px solid green;"> Line Chart

```python
plt.figure(figsize=(15,5))
plt.axes(facecolor='pink')
plt.plot(x, y,color='black',label='Line chart') #Line Chart
plt.xlim(0, 200) 
# plt.xticks(1)
plt.ylim(0, 10)
plt.xlabel('Sales',color='blue')
plt.ylabel('product')
plt.title('Line Plot')
plt.suptitle('Sales Comparison', size=20, y=1.05)
plt.savefig('Save.png')
plt.legend()
plt.show()
```

```python
import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/Apress/data-analysis-and-visualization-using-python/master/Ch07/Salaries.csv')
df.head()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2)
ax[0,0].set_title('hello')
ax[0,0].plot(x,y) #line chart
ax[0,1].scatter(x,y) #scatter plot
ax[1,0].hist(x) #histogram
ax[1,1].boxplot(x)# boxplot
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax[0,0].set_title('Plot 1')
ax[0,0].plot(x,y)
ax[0,1].hist(df['salary'])
ax[1,0].boxplot(df.salary)
ax[1,1].hist(df.service)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2,figsize=(10,10))

ax[0,0].plot(x,y,color='green',linestyle='dashed')
ax[0,0].set_title('Line')
ax[1,0].set_title('Scatter Plot')
ax[1,0].scatter(x,y)
ax[1,1].hist(y,color='orange')
ax[0,1].set_title('Piechart')
ax[1,1].set_title('Histogram')
ax[0,1].pie(df['rank'].value_counts())
plt.show()
```
